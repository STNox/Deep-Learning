# 공공 인공지능 Open API 활용

객체 검출(Object Detection)

```python
with open('etri_ai_key.txt') as kfile:
    eai_key = kfile.read(100)
```

```python
import json, base64
openApiURL = 'http://aiopen.etri.re.kr:8000/ObjectDetect'
```

```python
import os
image_file = './data/giraffe.jpg'
_, image_type = os.path.splitext(image_file)
image_type = 'jpg' if image_type == '.jfif' else image_type[1:]
```

```python
with open(image_file, 'rb') as file:
    image_contents = base64.b64encode(file.read()).decode('utf8')
```

```python
request_json = {
    'request_id': 'reserved field',
    'access_key': eai_key,
    'argument': {
        'file': image_contents,
        'type': image_type
    }
}
```

```python
import urllib3
http = urllib3.PoolManager()
response = http.request(
    'POST',
    openApiURL,
    headers={'Content-Type': 'application/json; charset=UTF-8'},
    body=json.dumps(request_json)
)
```

```python
print('[responseCode]', response.status)
```

```python
response.data
```

```python
import json
result = json.loads(response.data)
result
```

```python
obj_list = result['return_object']['data']
obj_list
```

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = Image.open(image_file)
plt.imshow(image);
```

```python
draw = ImageDraw.Draw(image)
for obj in obj_list:
    name = obj['class']
    x = int(obj['x'])
    y = int(obj['y'])
    w = int(obj['width'])
    h = int(obj['height'])
    draw.text((x+10, y+10), name, font=ImageFont.truetype('malgun.ttf', 20), fill=(255, 0, 0))
    draw.rectangle(((x, y), (x+w, y+h)), outline=(255, 0, 0), width=3)
```
